```python
# we are replacinf the missing value with the median as it is the best approach than mean
numeric_cols = [

    "trestbps",
    "chol",
    "thalch",
    "oldpeak"
]
for col in numeric_cols:
  df[col] = df[col].fillna(df[col].median())
```

```python
categorical_cols = ["fbs", "restecg", "exang", "slope", "ca", "thal"]

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
#now we are going to do encoding of the data eg changing the object data to numerical data as we  will get error while predicting the values and scikitlearn void this type of things
df["sex"].unique()
```

```python
df["sex"] = df["sex"].map({
    "Male": 1,
    "Female": 0
})
```
